Video Game Sales and Ratings.

This dashboard provides a comprehensive analysis of video game sales and ratings. With the increasing popularity of video games as a form of entertainment and the rapid growth of the gaming industry, understanding the trends and patterns in video game sales and ratings is crucial for both gamers and industry professionals.

```python
#necessary libraries
import pandas as pd
import numpy as np
```

```python
#import the dataset
df = pd.read_csv("vgsales.csv")
```

```python
#checking null values 
df.isnull().sum()
```

```python
#deleting all null values
data = df.dropna()
print(data)
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
for column in data:
    unique_vals = np.unique(data[column])
    nr_values = len(unique_vals)
    if nr_values < 13:
        print('The number of values for feature {} :{} -- {}'.
              format(column, 
        nr_values,unique_vals))
    else:
        print('The number of values for feature {} :{}'.
              format(column, nr_values))
```

```python
filtered_data = data[(data['Year'] >= 2000) & (data['Year'] <= 2016)]
filtered_data
```

```python
filtered_data.to_csv("video_games_new.csv", index=False)
```
